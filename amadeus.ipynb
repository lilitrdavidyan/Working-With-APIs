```python
from amadeus import Client
import spotipy
import requests
import time
import json
import pandas as pd
```

```python
secrets_file = open("secrets.txt","r")
```

```python
string = secrets_file.read()
```

```python
secrets_dict={}
for line in string.split('\n'):
    if len(line) > 0:
        #print(line.split(':'))
        secrets_dict[line.split(':')[0]]=line.split(':')[1].strip()
```

```python
amadeus = Client(
    client_id=secrets_dict['apikey'],
    client_secret=secrets_dict['apisecret']
)
```

```python
response = amadeus.reference_data.locations.hotels.by_city.get(cityCode='BER')
len(response.data)
```

```python
df = pd.DataFrame(response.data)
df
```
